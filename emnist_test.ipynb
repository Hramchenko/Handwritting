```python
classes = {}

for data, target in emnist:
    idx = target.item()
    if idx not in classes:
        classes[idx] = []
    classes[idx].append(data)
```

```python
classes.keys()
```

```python
for k in classes.keys():
    print("---------------------")
    print("Class : %d" % k)
    
    fig = plt.figure()
    v = classes[k]
    for i in range(0, 5):
        ax = fig.add_subplot(5,1,i+1)
        ax.imshow(v[i][0], cmap="gray")
    plt.show()
        
```
